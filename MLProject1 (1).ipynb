```python
### Step 1: Loading and Inspecting the Dataset
import pandas as pd

# Load the dataset
df = pd.read_csv('C:\\Users\\apoor\\Downloads\\Covid Data.csv\\Covid Data.csv')

# Displaying the first few rows and info about the dataset to understand its structure
df.head()
df.info()
#Checking the summary statistics
df.describe()
```

```python
#Step 2 : Data cleaning and preprocessing
# Converting 'DATE_DIED' to a boolean to indicate whether the patient died
df['DIED'] = df['DATE_DIED'] != '9999-99-99'

# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Apply median only to numeric columns and fill missing values
numeric_cols = df.select_dtypes(include='number').columns
df_cleaned = df.copy()

# Filling missing values in numeric columns with median
df_cleaned[numeric_cols] = df_cleaned[numeric_cols].fillna(df[numeric_cols].median())

# For non-numeric columns, you can fill missing values with a mode or a specific value
non_numeric_cols = df.select_dtypes(exclude='number').columns

# Check the cleaned data info
df_cleaned.info()
```

```python
# Step 3: Exploratory Data Analysis (EDA)

## Gender and Age Distribution
import matplotlib.pyplot as plt
import seaborn as sns

# Gender distribution
plt.figure(figsize=(6, 5))
sns.countplot(x='SEX', data=df)
plt.title('Gender Distribution (1 = Male, 2 = Female)')
plt.show()

# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

## Comorbidities and Outcomes
# Relevant columns for comorbidities and outcomes
comorbidity_columns = ['DIABETES', 'OBESITY', 'ICU', 'CLASIFFICATION_FINAL']
df_comorbidity = df[comorbidity_columns]

# Diabetes distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='DIABETES', data=df_comorbidity)
plt.title('Diabetes Distribution (1 = Yes, 2 = No)')
plt.show()

# Obesity distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='OBESITY', data=df_comorbidity)
plt.title('Obesity Distribution (1 = Yes, 2 = No)')
plt.show()
```

```python
### Step 4: Mortality Analysis
# Boxplot for age vs. mortality
plt.figure(figsize=(10, 6))
sns.boxplot(x='DIED', y='AGE', data=df)
plt.title('Age and Mortality')
plt.show()

# Mortality and comorbidities
plt.figure(figsize=(10, 6))
sns.countplot(x='DIABETES', hue='DIED', data=df, palette='Set1')
plt.title('Diabetes and Mortality')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='OBESITY', hue='DIED', data=df, palette='Set2')
plt.title('Obesity and Mortality')
plt.show()
```

```python
### Step 5: ICU Admission Analysis
# ICU admissions by comorbidity
plt.figure(figsize=(10, 6))
sns.countplot(x='DIABETES', hue='ICU', data=df)
plt.title('ICU Admissions and Diabetes')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='OBESITY', hue='ICU', data=df)
plt.title('ICU Admissions and Obesity')
plt.show()

```

```python
### Step 6: Statistical Inference
from scipy import stats

# Age comparison between ICU and non-ICU patients
icu_patients = df[df['ICU'] == 1]['AGE']
non_icu_patients = df[df['ICU'] == 2]['AGE']

# Perform t-test
t_stat, p_value = stats.ttest_ind(icu_patients, non_icu_patients, nan_policy='omit')
print(f"T-statistic: {t_stat}, P-value: {p_value}")

if p_value < 0.05:
    print("There is a significant difference in age between ICU and non-ICU patients.")
else:
    print("There is no significant difference in age between ICU and non-ICU patients.")
```

# Step 7 conclusion 
This analysis revealed significant trends:
- Older patients were more likely to be admitted to the ICU.
- Comorbidities such as diabetes and obesity correlated with higher mortality rates.
- Statistical testing showed a significant difference in age between ICU and non-ICU patients.
